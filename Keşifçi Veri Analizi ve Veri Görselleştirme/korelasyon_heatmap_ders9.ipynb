# Korelasyon - Heatmap

```python
import seaborn as sns 
flights = sns.load_dataset('flights')
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df = df.pivot("month","year","passengers")
df.head()
```

```python
# Yılların ve ayların birbirleri arasındaki ilişkiyi gösterebilmek için ısı haritası kullanılır.
# Isı haritası oluşturmak için heatmap kullanılır.
sns.heatmap(df);
```

```python
# Heatmap pattern detection için bir çok yerde kullanılır.
```

```python
# Ölçek değerleri göstererek görselleştirme
sns.heatmap(df, annot = True, fmt = 'd');
```

```python
# Grafiği daha okunur bir hale getirmek için
sns.heatmap(df, annot = True, fmt = 'd', linewidths = .5);
```

```python
sns.heatmap(df, cbar = False); # Renk paletini kaldırmak için
```

```python
# Çıktı boyutunu ayarlamak için
sns.set_context("poster")
sns.set(rc={'figure.figsize': (10, 10)})
sns.set_style("whitegrid")
```
