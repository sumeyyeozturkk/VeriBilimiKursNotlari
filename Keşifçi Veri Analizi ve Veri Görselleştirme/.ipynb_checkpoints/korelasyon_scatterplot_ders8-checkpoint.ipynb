# Korelasyon - Scatterplot

Korelasyon --> Değişkenlerin birbirleri arasındaki ilişkiyi ifade eden bir tekniktir.

```python
import seaborn as sns

iris = sns.load_dataset('iris')
df = iris.copy()
df.head()
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue = 'species');
```

```python
sns.pairplot(df, hue = 'species', markers = ['o', 's', 'D']);
```

```python
# alt kümelerini çizdirme
sns.pairplot(df, vars = ['sepal_length','sepal_width']);
```

```python
sns.pairplot(df,vars = ['sepal_length','sepal_width'], hue= 'species', kind = 'reg'); 
# Regresyon doğrusu ekleyerek korelasyonu daha iyi gözlemleyebiliriz.
```

# Bağımlı Yoğunluk Grafiği ve Scatterplot Matrisi

```python
#Kodların aynı hüsrede olmasına dikkat edilmeli.

g = sns.PairGrid(df, diag_sharey = False) #Hücreleri oluşturma
g.map_lower(sns.kdeplot) # Hücre özelliklerini oluşturma
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot, lw = 3)
```
