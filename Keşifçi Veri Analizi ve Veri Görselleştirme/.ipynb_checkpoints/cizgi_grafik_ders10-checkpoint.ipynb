```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
df['timepoint'].describe()
```

```python
df['signal'].describe()
```

```python
sns.jointplot(x = 'timepoint', y = 'signal', data = df, kind = 'reg');
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

```python
sns.lineplot(x = "timepoint",  y = "signal", data = df)
# Lineplot kullanarak kullanmÄ±ÅŸ olduÄŸumuz veri setinin iÃ§erisinde bulunan her bir timepoint noktasÄ±ndaki 
#signal deÄŸerlerinin ortamasÄ±nÄ± Ã§izdiriyoruz.
#EtrafÄ±ndaki soluk yerler de standart sapmasÄ±nÄ± ifade ediyor.
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = df)
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style = "event", data = df)
```

```python
sns.lineplot(x = "timepoint", y = "signal", 
             hue = "event", style = "event",
             markers = True, dashes = False, data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "region", 
             style = "event", data = df);
```
