# Violin

Boxplot'tan farkı köşelerinin biraz daha kırılgan ve hassas yapıda olmasıdır.
Aynı zamanda basıklık ile ilgili bilgi de verir.

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.catplot(y = "total_bill", kind = "violin", data = df);
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', data = df);
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', hue = 'sex', data = df);
```

```python
# FacetGrid üzerinde bazı işlemler
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(df, row = "sex", col = "day", margin_titles = True) 
#bolme işlemi

bins = np.linspace(0, 60, 13) #Histogramın aralıkları


g.map(plt.hist, "total_bill", color="steelblue", bins=bins);
```
