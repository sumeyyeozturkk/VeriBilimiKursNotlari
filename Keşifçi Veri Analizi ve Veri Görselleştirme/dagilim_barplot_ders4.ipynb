# Değişkenlerin Dağılımlarını Gösterme Teknikleri

# Barplot

Diamonds veri seti:

price: dolar cinsinden fiyat ($326 $18,823)

carat: ağırlık (0.2–5.01)

cut: kalite (Fair, Good, Very Good, Premium, Ideal)

color: renk (from J (worst) to D (best))

clarity: temizliği, berraklığı 
    (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

x: length in mm (0–10.74)

y: width in mm (0–58.9)

z: depth in mm (0–31.8)

depth: toplam derinlik yuzdesi = z / mean(x, y) = 2 * z / (x + y) (43–79)

table: width of top of diamond relative to widest point (43–95)

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe().T
```

```python
from pandas.api.types import CategoricalDtype
```

```python
# Sıralı Nominal Değerleri Ordinal Kategorik değişken türüne dönüştürme
df.cut = df.cut.astype(CategoricalDtype(ordered = True))
df.color = df.color.astype(CategoricalDtype(ordered = True))
df.clarity = df.clarity.astype(CategoricalDtype(ordered = True))
```

```python
# Kategorik değişkenin ordinal ölçet türüne dönüştürülmesi yanlış gerçekleşmiş.

df.cut.head(1)
# cut: kalite (Fair, Good, Very Good, Premium, Ideal)
```

```python
df.color.head(1)
# renk (from J (worst) to D (best))
```

```python
df.clarity.head(1)
# clarity: temizliği, berraklığı (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
```

```python
# Bu yanlışlığı düzeltmemiz gerekiyor. Bunu şu şekilde gerçekleştirebiliriz.
```

```python
cat_kategoriler = ['Fair','Good','Very Good', 'Premium', 'Ideal']
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories = cat_kategoriler, ordered = True))
```

```python
df.cut.head(1)
```

# Barplot

NOT : Barplot Kategorik değişken görselleştirmek için kullanılır.

```python
(df['cut']
 .value_counts()
 .plot.barh()
 .set_title('Barplot Kategorik Değişken Görselleştirmek İçin Kullanılır.'));
```

NOT : Bu kategorik değişkenin veri setine olan etkinisi görmek için diğer değişkenler ile birlikte görselleştirilmesi gerekir. 

```python
# catplot() --> Kategorik değişken görselleştirme için optimize edilmiş fonk.
sns.catplot(x = 'cut', y = 'carat', data = df);
```

```python
sns.catplot(x = 'cut', y = 'price', data = df);
# Diğer değişkenler ile çarpıştırma işlemi
```

```python
%config InlineBackend.figure_format = 'retina' # Görüntünün kalitesi için yaptığımız bir ayar.
sns.catplot(x = 'cut', y = 'price', hue = 'color', data = df);
# Bir katman daha ekleyerek veri setini daha iyi analiz edebilme durumunu sağlıyoruz.
```
