# DataFrame Birleştirme İşlemleri

```python
#Birleştirme(Join İşlemleri)
```

```python
import numpy as np
import pandas as pd
```

```python
s1 = np.random.randint(10, size = 5)
s2 = np.random.randint(10, size = 5)
s3 = np.random.randint(10, size = 5)

df1 = pd.DataFrame({'var1': s1, 'var2': s2, 'var3': s3})
df1
```

```python
df2 = df1 + 99
df2
```

```python
pd.concat([df1,df2]) # Satır bazında concat ile birleştirme işlemi
```

```python
pd.concat([df1,df2], axis = 1) # Sütun bazında concat ile birleştirme işlemi
```

```python
# Satır bazında gerçekleştirilen birleştirme işlemi indexlerde çoklama problemine neden olur.
```

```python
pd.concat([df1,df2], ignore_index = True) # Bu argüman ile çoklama problemini ortadan kaldırıyoruz.
```

```python
# DataFrame'in değişken isminin farklı olduğu durumda yapılacaklar
```

```python
df2.columns = ['var1','var2','deg3']
df2
```

```python
df1
```

```python
pd.concat([df1,df2]) # Bu birleştirmede aynı isimli olmayan değişkenlerin değerlerini 'NaN' olarak yazıyor
```

```python
# Sadece kesişen değerleri birleştirerek 'NaN' ifadelerini görmüyoruz.
pd.concat([df1,df2], join = 'inner')
```

```python
# Sadece bir df'nin değişken isimlerini referans alarak birleştirme işlemi gerçekleştirilir.
pd.concat([df1,df2], join_axes = [df1.columns]) 
```

# Merge ve Join İşlemleri

```python
df1 = pd.DataFrame({'calisanlar': ['Ali', 'Veli', 'Ayse', 'Fatma'],
                    'grup': ['Muhasebe', 'Muhendislik', 'Muhendislik', 'İK']})

df1
```

```python
df2 = pd.DataFrame({'calisanlar': ['Ayse', 'Ali', 'Veli', 'Fatma'],
                    'ilk_giris': [2010, 2009, 2014, 2019]})

df2
```

```python
# One to one Birleştirme İşlemi
```

```python
pd.merge(df1,df2) # Çalışanlara göre birleştirdi.
```

```python
pd.merge(df1,df2, on = 'calisanlar') # Özellikle bir değişkene göre birleştirmek istediğimizde
```

```python
# Many to One Birleştirme işlemi 
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({'grup': ['Muhasebe', 'Muhendislik', 'İK'],
                    'mudur': ['Caner', 'Mustafa', 'Berkcan']})

df4
```

```python
pd.merge(df3,df4)
```

```python
# Many to Many Birleştirme İşlemi
```

```python
df5 = pd.DataFrame({'grup': ['Muhasebe', 'Muhasebe',
                              'Muhendislik', 'Muhendislik', 'İK', 'İK'],
                    'yetenekler': ['matematik', 'excel', 'kodlama', 'linux',
                               'excel', 'yonetim']})

df5
```

```python
df1
```

```python
pd.merge(df1,df5) # Örneğin Ali'nin iki farklı yeteneğini göstermek için çoklama yapıldı. 
```

```python
df3 = pd.DataFrame({'name': ['Ali','Veli','Ayşe','Fatma'],
                   'maas': [70000,80000,120000,90000]})
df3
```

```python
df1
```

```python
pd.merge(df1,df3) 
# Bu hatayı almamızın sebebi birleştirilecek ortak bir değişken ismi bulunmamasıdır.
#Ancak değişken isimleri farklı olmasına rağmen içerik olarak aynı olan değişkenimiz var.
```

```python
pd.merge(df1,df3, left_on = 'calisanlar', right_on = 'name') 
# Bu argümanlar ile iki değişkeni de kullanarak birleşmeyi sağlayabiliriz.
#Ancak değişken isimlerinde bir çoklama oluşmaktadır.
```

```python
pd.merge(df1,df3, left_on = 'calisanlar', right_on = 'name').drop('name', axis = 1) 
#Drop ifadesini kullanarak içlerinden silmek istediğimizi silebiliriz. 
#Birleştirmeden sonra sadece tek bir değişken kalacaktır.
```

```python
## İndexlere Göre Birleştirme İşlemleri
```

```python
df1a = df1.set_index('calisanlar') # Bu değişkenimiz artık index'imiz oluyor.
df1a
```

```python
df2a = df2.set_index('calisanlar')
df2a
```

```python
pd.merge(df1a,df2a, left_index = True, right_index = True) # İndexlere göre birleştirme işlemi
```

```python
#Bunun yerine join de kullanabiliriz. Ön tanımlı olarak indexe göre birleştirme yapıyor.
df1a.join(df2a)
```

## Overlapping problemi

```python
# İki DataFrame birleştirdiğimizde kesişen ilk ismi merge fonk. birleştirme değişkeni olarak kabul eder.
# Fakat bundan başka  değişken ismi kesişirse
#bu değişkenin isimlerini verisetinde tutar ama kendisi isimlendirerek tutar.
```

```python
dfa = pd.DataFrame({'calisanlar': ['Ali', 'Veli', 'Ayse', 'Fatma'],
                    'siralama': [1, 2, 3, 4]})

dfa
```

```python
dfb = pd.DataFrame({'calisanlar': ['Ali', 'Veli', 'Ayse', 'Fatma'],
                    'siralama': [3, 1, 4, 2]})

dfb
```

```python
pd.merge(dfa,dfb, on = 'calisanlar') 
#'Calisanlar' değişken isminden başka kesişen bulduğu için bunları kendisi isimlendirerek görüntüledi.
```

```python
pd.merge(dfa,dfb, on = 'calisanlar', suffixes = ['_MAAS','_DENEYIM']) 
# Bu argüman ile kendimiz isimlendirerek probleme çözüm getirebiliyoruz.
```
