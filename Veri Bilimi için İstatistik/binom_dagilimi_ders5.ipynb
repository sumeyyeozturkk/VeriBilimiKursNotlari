# Binom Dağılımı

Binom dağılımı, n deneme sonucu k başarılı olma olasılığı ile ilgilenildiğinde kullanılan bir dağılımdır.

Deneyler bağımsız ve aynı koşullar altında gerçekleştiriliyor.

```python
from scipy.stats import binom
import seaborn as sns

veri_binom = binom.rvs(n = 15, p = 0.5, size = 10000)

ax = sns.distplot(veri_binom,
                  kde=True,
                  color='orange',
                  hist_kws={"linewidth": 15,'alpha':1})

ax.set(xlabel = 'Binom Dağılımı', ylabel = 'Frekans')

#kde True olduğunda olasılık değerlerini gösteren bir grafik oluşturuyor.
```

```python
import scipy.stats as stats
from matplotlib.pyplot import plot
import numpy as np
```

```python
bd1 = stats.binom(20, 0.5)
bd2 = stats.binom(20, 0.7)
bd3 = stats.binom(40, 0.5)

k = np.arange(40)

plot(k, bd1.pmf(k), 'o-b')
plot(k, bd2.pmf(k), 'd-r')
plot(k, bd3.pmf(k), 's-g')
```
