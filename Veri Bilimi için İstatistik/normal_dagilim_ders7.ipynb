# Normal Dağılım

Normal dağıldığı bilinen sürekli rassal değişkenler için olasılık hesaplamasında kullanılır.

Günlük hayatta karşılaştığımız bir çok olayın olasılık yoğunluk fonksiyonu aritmetik ortalama etrafında yüksek ve uçlara doğru azalan bir yapıda olmaktadır. Bu yapıya normal dağılım(çan eğrisi) denir. Makine öğrenmesinde genelde Gauss dağılımı olarak adlandırılır.

```python
from scipy.stats import norm
import scipy.stats as stats
import seaborn as sns
import numpy as np
from matplotlib.pyplot import plot


veri_normal = norm.rvs(size=10000,loc=0,scale=1)

ax = sns.distplot(veri_normal,
                  bins=100,
                  kde=False,
                  color='orange',
                  hist_kws={"linewidth": 15,'alpha':1})

ax.set(xlabel='Normal', ylabel='Frekans')
```

```python
k = np.linspace(-10,10)
pd = stats.norm(0,1)
plot(k, pd.pdf(k))
```

```python
nm1 = stats.norm(20,5)
nm2 = stats.norm(10, 2)
nm3 = stats.norm(30, 7)
k = np.arange(40)
plot(k, nm1.pdf(k), 'o-b')
plot(k, nm2.pdf(k), 'd-r')
plot(k, nm3.pdf(k), 's-g')
```
